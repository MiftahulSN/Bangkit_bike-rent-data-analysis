### Explore <span style="background-color: lightgray;">day.csv</span>

```python
byDay_df.describe(include="all")
```

```python
season_name = {
    1: 'Springer',
    2: 'Summer',
    3: 'Fall',
    4: 'Winter'
}

byDay_df['season'] = byDay_df['season'].replace(season_name)

result = byDay_df.groupby('season').agg({
    'cnt': 'sum', 
    'temp': ['mean','std','min','max'], 
    'hum': ['mean','std','min','max'], 
    'windspeed': ['mean','std','min','max']
})

result.columns = ['_'.join(col).strip() for col in result.columns.values]
result = result.reset_index()

result.sort_values(by='cnt_sum', ascending=False)
```

```python
weather_name = {
    1: 'Clear, Cloudy',
    2: 'Mist, Broken Cloud',
    3: 'Light Snow, Light Rain',
    4: 'Heavy Rain, Thunderstorm'
}

byDay_df['weathersit'] = byDay_df['weathersit'].replace(weather_name)

result = byDay_df.groupby('weathersit').agg({
    'cnt': 'sum', 
    'temp': ['mean','std','min','max'], 
    'hum': ['mean','std','min','max'], 
    'windspeed': ['mean','std','min','max']
})

result.columns = ['_'.join(col).strip() for col in result.columns.values]
result = result.reset_index()

result.sort_values(by='cnt_sum', ascending=False)
```

#### Insight : 
+ pada *exploratory data* hour.csv, dilakukan pengelompokan jumlah penyewaan sepeda berdasarkan jam, dengan total penyewaan dibagi antara pengguna umum ('casual') dan yang terdaftar member ('registered'). hasil diurutkan dari yang tertinggi ke terendah berdasarkan total penyewaan ('cnt'). terlihat bahwa jam 17:00 atau pukul 05:00 PM merupakan waktu penyewaan sepeda paling ramai dan pada jam 04:00 atau pukul 04:00 AM merupakan waktu penyewaan sepeda paling sepi  
+ pada *exploratory data* day.csv, dilakukan pengelompokan jumlah penyewaan sepeda berdasarkan musim dan cuaca. hasil menunjukkan bahwa *fall season* atau musim gugur merupakan musim penyewaan sepeda paling ramai dan *springer season* atau musim semi merupakan musim penyewaan sepeda paling sepi. juga pada cuaca cerah berawan atau *clear or cloudy weather* merupakan cuaca dengan penyewaan sepeda paling ramai dan pada cuaca bersalju atau gerimis *light snow or light rain* merupakan cuaca dengan penyewaan sepeda paling sedikit atau sepi.

## Visualization & Explanatory Analysis

### Question 1 :

```python
time_segments = {
    'Early Morning [02:00 AM - 04:59 AM]': (2, 4),
    'Morning [05:00 AM - 10:59 AM]': (5, 10),        
    'Afternoon [11:00 AM - 02:59 PM]': (11, 14),      
    'Evening [03:00 PM - 05:59 PM]': (15, 17),       
    'Night [06:00 PM - 09:59 PM]': (18, 21),          
    'Midnight [10:00 PM - 01:59 AM]': (22, 1)         
}

def segmentation(hour):
    for segment, (start, end) in time_segments.items():
        if start <= hour <= end or (start > end and (hour >= start or hour <= end)):
            return segment
    return None

byHour_df['time_segment'] = byHour_df['hr'].apply(segmentation)

bike_rent = byHour_df.groupby('time_segment')['cnt'].sum()
bike_rent = bike_rent.sort_values(ascending=True)

plt.figure(figsize=(10, 6))
bars = bike_rent.plot(kind='barh', color='skyblue')

for bar in bars.patches:
    plt.text(
        bar.get_width(),
        bar.get_y() + bar.get_height() / 2,
        int(bar.get_width()),
        va='center', 
        ha='left',
        color='red'
    )

plt.title('Bike Rentals by Time Segment in 2 Years (2011 - 2012)')
plt.xlabel('Total Bike Rentals (in Unit)')
plt.ylabel('Time Segment (in A Day)')
plt.grid(axis='x')
plt.tight_layout()
plt.show()
```

### Question 2 :

```python
season_weather_counts = byDay_df.groupby(['season', 'weathersit'])['cnt'].sum().reset_index()

rent_by_season = season_weather_counts.groupby('season')['cnt'].transform('sum')

season_weather_counts['percentage'] = (season_weather_counts['cnt'] / rent_by_season) * 100

pivot_table = season_weather_counts.pivot(index='season', columns='weathersit', values='percentage')

fig, ax = plt.subplots(figsize=(10, 6))  

pivot_table.plot(kind='bar', stacked=True, 
                 color=['#ff9999', '#66b3ff', '#99ff99', '#ffcc99'], ax=ax)

plt.title('Percentage of Bike Rentals by Season and Weather')
plt.xlabel('Season')
plt.ylabel('Percentage of Bike Rentals (%)')

plt.xticks(rotation=0) 
plt.legend(title='Weather Situation', 
           labels=['Clear', 'Light Rain', 'Mist + Cloudy', 'Heavy Rain'])
plt.grid(axis='y')

for i, total in enumerate(rent_by_season.drop_duplicates()): 
    ax.text(i, 50, str(int(total)), 
            ha='center', 
            va='center', 
            color='white', 
            fontsize=10, 
            fontweight='bold')

plt.tight_layout()
plt.show()
```

#### Insight : 
+ grafik visualisasi data hour.csv menunjukkan segmen waktu tersibuk atau jumlah penyewaan sepeda terbanyak adalah pada malam hari (06:00 PM - 09:59 PM). hal ini menunjukkan bahwa banyak orang menyewa sepeda pada malam hari, mungkin untuk rekreasi atau aktivitas lainnya. sementara segmen waktu paling sepi atau jumlah penyewaan sepeda paling sedikit adalah dini hari (02:00 AM - 04:59 AM). hal ini masuk akal karena umumnya aktivitas di luar ruangan pada dini hari sangat minim.
+ grafik visualisasi data day.csv menunjukkan cuaca cerah (*clear*) sangat dominan di setiap musim. hal ini mungkin karena cuaca cerah memiliki suhu dan kondisi yang nyaman untuk bersepeda. secara umum, cuaca cerah (merah muda) memiliki persentase penyewaan sepeda tertinggi di semua musim. selain itu, musim gugur (*fall season*) memiliki jumlah penyewaan sepeda tertinggi secara keseluruhan, diikuti oleh musim panas (*summer*). hal ini mungkin karena faktor-faktor seperti suhu yang nyaman dan lebih banyaknya aktivitas luar ruangan pada musim-musim tersebut.

## Advanced Analytics

### Clustering Analytics

```python
def windspeed_cluster(windspeed):
    if 0 <= windspeed <= 0.33:
        return 'LOW'
    elif 0.34 <= windspeed <= 0.66:
        return 'MEDIUM'
    else:
        return 'HIGH'

byHour_df['windspeed_cluster'] = byHour_df['windspeed'].apply(windspeed_cluster)


windspeed_groups = byHour_df.groupby('windspeed_cluster')['cnt'].sum().reset_index()

plt.figure(figsize=(10, 6))
bars = plt.bar(windspeed_groups['windspeed_cluster'], windspeed_groups['cnt'], color=['#66b3ff', '#ff9999', '#99ff99'])

for bar in bars:
    yval = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, yval, f'{yval:,}', 
             ha='center', va='bottom')  

plt.title('Total Bike Rentals by Windspeed Cluster (hour.csv dataset)')
plt.xlabel('Windspeed Cluster')
plt.ylabel('Total Rentals (cnt)')
plt.xticks(rotation=45)  
plt.tight_layout()
plt.show()
```

```python
def windspeed_cluster(windspeed):
    if 0 <= windspeed <= 0.33:
        return 'LOW'
    elif 0.34 <= windspeed <= 0.66:
        return 'MEDIUM'
    else:
        return 'HIGH'

byDay_df['windspeed_cluster'] = byDay_df['windspeed'].apply(windspeed_cluster)


windspeed_groups = byDay_df.groupby('windspeed_cluster')['cnt'].sum().reset_index()

plt.figure(figsize=(10, 6))
bars = plt.bar(windspeed_groups['windspeed_cluster'], windspeed_groups['cnt'], color=['#66b3ff', '#ff9999', '#99ff99'])

for bar in bars:
    yval = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, yval, f'{yval:,}', 
             ha='center', va='bottom')  

plt.title('Total Bike Rentals by Windspeed Cluster (day.csv dataset)')
plt.xlabel('Windspeed Cluster')
plt.ylabel('Total Rentals (cnt)')
plt.xticks(rotation=45)  
plt.tight_layout()
plt.show()
```

#### Insight : 
+ berdasarkan plot grafik dari kedua dataset menunjukkan bahwa kecepatan angin menjadi faktor yang sangat mempengaruhi penyewaan secara signifikan. kondisi angin rendah memiliki total penyewaan sepeda tertinggi dibandingkan dengan kondisi angin sedang dan tinggi. hal ini menunjukkan bahwa kondisi angin kencang dan sedang mengurangi minat masyarakat untuk bersepeda.
